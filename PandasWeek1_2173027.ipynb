**#01 : Menyertakan Prefix dan Suffix pada seluruh kolom Data Frame**

**Import Modules**

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

**Persiapan Data Frame**

```python
n_rows = 5
n_cols = 5
cols = tuple ('ABCDE')

df = pd.DataFrame(np.random.randint(1,10, size = (n_rows, n_cols)),
                  columns = cols)
df
```

```python
tuple('ABCDE')
```

```python
df.add_prefix('kolom_')
```

Penggunaan Prefix adalah untuk menambahkan String pada bagian depan pada string tertentu

**Menyertakan Suffix Kolom**

```python
df.add_suffix('_field')
```

Penggunaan Suffix adalah untuk menambah String di bagian belakang

**#02 : Pemilihian baris ( rows selection ) pada Data Frame**


**Persiapan Data Frame**

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1,5, size = (n_rows, n_cols)),
                  columns = cols)
df
```

**Selection dengan operator logika |**

```python
df[(df['A'] == 1) | (df['A'] == 3)]
```

**Selection dengan fungsi isin()**

```python
df[df['A'].isin([1,3])]
```

**Mengenal operator negasi ~**

operator negasi dimanfaatkan atau digunakan untuk mengenali operator not


```python
df[~df['A'].isin([1,3])]
```

**#03 : Konversi tipe data String ke Numerik pada kolom Data Frame**


**Import Modules**

```python
import pandas as pd
print(pd.__version__)
```

**Persiapan Data Frame**

```python
data = {'col1' : ['1', '2', '3', 'teks'],
        'col2' :['1', '2', '3', '4']}

df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

**Konversi tipe data dengan fungsi astype()**

```python
df_x = df.astype({'col2' : 'int'})
df_x
```

```python
df_x.dtypes
```

**Konversi tipe data numerik dengan fungsi to_numeric()**

```python
df.apply(pd.to_numeric, errors ='coerce')
```

**#04 : Pemilihan kolom ( columns selection ) pada Data Frame berdasarkan tipe data**


**Import Modules**

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

**Persiapan Data Frame**

```python
n_rows = 5
n_cols = 2
cols = ['bil_pecahan', 'bil_bulat']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns=cols)
df['bil_pecahan'] = df['bil_pecahan'].astype('float') #--> Hal ini akan mengkonversi pecahan menjadi tipe data float

df.index = pd.util.testing.makeDateIndex(n_rows, freq= 'H') 
df = df.reset_index()

df['teks'] = list('ABCDE')

df
```

```python
df.dtypes
```

**Memilih kolom bertipe data numerik**

```python
df.select_dtypes(include='number')
```

```python
df.select_dtypes(include='float')
```

```python
df.select_dtypes(include='int')
```

**Memilih kolom bertipe data string atau object**

```python
df.select_dtypes(include='object') #--> fungsi dari ini adalah untuk memanggil kolom yang memiliki data tentunya numerik
```

**Memilih kolom bertipe data string atau datetime**

```python
df.select_dtypes(include='datetime') #fungsi dari ini adalah untuk memanggil kolom yang memiliki tipe data "DATE"
```

**Memilih kolom dengan kombinasi tipe data**

```python
df.select_dtypes(include=['number', 'object'])
```

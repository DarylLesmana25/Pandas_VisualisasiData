# #07 Menghapus (drop) Missing Values (NaN) pada Data Frame
Import Modules

```python
import pandas as pd

print(pd.__version__)
```

```python
df = pd.util.testing.makeMissingDataframe().reset_index() 
df.head()
```

```python
df = df.rename(columns={'index':'Z'})
df.head()
```

```python
df_backup = df.copy(deep=True)
```

Menghapus ( drop ) setiap Kolom yang Mengandung Missing Values

```python
df = df.dropna(axis='columns') 
df.head() 
```

Menghapus ( drop ) Setiap Baris yang Mengangund Missing Values

```python
df = df_backup.copy(deep = True)

df = df.dropna(axis = 'rows')
df.head()
```

Persentase Missing Values untuk tiap Kolom

```python
df = df_backup.copy(deep=True)

df.isna().mean()*100
```

Menghapus (drop) Setiap Kolom yang Mengandung Missing Values Berdasarkan Threshold ( ambang batas )

```python
treshold = len(df) * 0.9 # len() digunakan untuk menghitung panjang baris dan variable thershold yang berisi batas 0.9 atau 90%

df = df.dropna(thresh=treshold, axis='columns')
df.head()
```

```python
data = {'A':[15, 15, 18, np.nan, 12], 
        'B':[15, 15, 18, np.nan, 12]}

df = pd.DataFrame(data)
df
```
